```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivada_lateral(f, x, h):
  return (f(x+h) - f(x))/h
```

```python
def derivada_central(f, x, h):
  return (f(x+h/2) - f(x-h/2))/h
```

## Gráfico do erro em função de h
### Varios pontos e várias funções diferentes

```python
def plot_erro(f, df, a, b, d_type=derivada_lateral):    
    hs = np.linspace(1e-2,1e-9,100)
    plt.figure(figsize=(10,7))

    for x in np.linspace(a, b,10):
        ys = abs( d_type(f, x, hs) - df(x) )
        plt.plot(hs, ys, label=f"x={x}")

    plt.legend()
    plt.grid()
    plt.xlabel("h")
    plt.ylabel("erro");
```

```python
plot_erro(np.sin, np.cos, 0, np.pi*2, derivada_central)
```
