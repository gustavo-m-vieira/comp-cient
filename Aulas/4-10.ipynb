```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivada_lateral(f, x, h):
  return (f(x+h) - f(x))/h
```

```python
def derivada_central(f, x, h):
  return (f(x+h/2) - f(x-h/2))/h
```

## Gráfico do erro em função de h
### Varios pontos e várias funções diferentes

```python
def plot_erro(f, df, a, b, d_type=derivada_lateral):    
    hs = np.logspace(-53, 0 , base=2, num=100)
    plt.figure(figsize=(10,7))

    for x in np.linspace(a, b,10):
        ys = abs( d_type(f, x, hs) - df(x) )
        plt.plot(hs, ys, label=f"x={x}")

    plt.legend()
    plt.grid()
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("h")
    plt.ylabel("erro");
```

```python
plot_erro(lambda x: x**4 + x**3 + 12, lambda x: 4*(x**3) + 3*(x**2), 0.1, 15, derivada_central)
```
